```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook, gridplot

output_notebook()
np.set_printoptions(formatter={"float": lambda x: f"{x:.1f}"})
```

# keep history of time

```python
ke = 200

ex = np.zeros(ke)
hy = np.zeros(ke)

# Pulse params

kc = ke // 2
t0 = 40
spread = 12

nsteps = 300

ex_hist = np.zeros([nsteps, ke])
hy_hist = np.zeros([nsteps, ke])


# FDTD loop
for ts in range(1, nsteps + 1):
    # Ex field
    for k in range(1, ke):
        ex[k] = ex[k] + 0.5 * (hy[k - 1] - hy[k])

    # gaussian pulse in the middle
    pulse = np.exp(-0.5 * ((t0 - ts) / spread) ** 2)
    ex[kc] = pulse

    # Hy field
    for k in range(ke - 1):
        hy[k] = hy[k] + 0.5 * (ex[k] - ex[k + 1])

    ex_hist[ts - 1, :] = ex
    hy_hist[ts - 1, :] = hy
```

```python
##############
_fparams = dict(width=900, height=300)
fig1 = figure(**_fparams)
fig2 = figure(**_fparams)

x = np.arange(len(ex))

_params = dict(level="image", x=0, y=0, dw=ke, dh=nsteps, palette="Turbo256")

fig1.image(image=[ex_hist], **_params)
fig2.image(image=[hy_hist], **_params)

show(gridplot([[fig1], [fig2]]))
```
